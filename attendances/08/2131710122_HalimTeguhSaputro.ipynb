### Soal 3 (skor 20/30)
Buatlah sebuah program dengan menggunakan **satu jenis perulangan** untuk menghitung nilai berdasarkan masukkan dari user. Perulangan akan terus berjalan jika user mengetikkan huruf kapital **"Y"** dan perulangan akan berhenti saat user mengetikkan huruf kapital **"N"**. 
Setelah perulangan berhenti, program akan memilih **angka terkecil** dan **angka terbesar** dari angka yang dimasukkan user dan menghitung jumlah dan rata-rata angkat-angka tersebut. Program boleh menggunakan salah satu perulangan : 

  + _While_ (20 poin), atau
  + _Do-while_ (20 poin), atau
  + _For_ (30 poin)
  
Contoh ![Contoh Soal 3](images/Soal3.png)


```java
/* Jawaban Soal 3 */
import java.util.Scanner;
Scanner input3 = new Scanner(System.in);
System.out.println("*** PROGRAM MENGHITUNG ANGKA ***");
System.out.println("NIM\t: 2131710122");
System.out.println("Nama\t: Halim Teguh Saputro");
System.out.println("---------------------------------------------");

System.out.println("\n-----------------FOR----------------------");
char tanya;
int angka, terkecil = 100000, terbesar, total, avg;
for(int i = 1; true; i++){
    System.out.print("masukkan angka ke-" + i + " : ");
    angka = input3.nextInt();
    total += angka;
    if(terbesar < angka){
        terbesar = angka;
    }
    if(terkecil > angka){
        terkecil = angka;
    }
    avg = total / i;
    System.out.print("Lanjut?");
    tanya = input3.next().charAt(0);
    if(tanya == 'N'){
        break;
    }
}
System.out.println("-----------------Hasil FOR----------------------");
System.out.println("Angka Terkecil : " + terkecil);
System.out.println("Angka Terbesar : " + terbesar);
System.out.println("Total bilangan : " + total);
System.out.println("rata-rata bilangan : " + avg);

System.out.println("\n-----------------WHILE----------------------");
int i = 1;
total = 0;
while(true){
    System.out.print("masukkan angka ke-" + i + " : ");
    angka = input3.nextInt();
    total += angka;
    if(terbesar < angka){
        terbesar = angka;
    }
    if(terkecil > angka){
        terkecil = angka;
    }
    avg = total / i;
    System.out.print("Lanjut?");
    tanya = input3.next().charAt(0);
    if(tanya == 'N'){
        break;
    }
    i++;
}

System.out.println("-----------------Hasil WHILE----------------------");
System.out.println("Angka Terkecil : " + terkecil);
System.out.println("Angka Terbesar : " + terbesar);
System.out.println("Total bilangan : " + total);
System.out.println("rata-rata bilangan : " + avg);

System.out.println("\n-----------------DO-WHILE----------------------");
int i = 1;
total = 0;
do{
    System.out.print("masukkan angka ke-" + i + " : ");
    angka = input3.nextInt();
    total += angka;
    if(terbesar < angka){
        terbesar = angka;
    }
    if(terkecil > angka){
        terkecil = angka;
    }
    avg = total / i;
    System.out.print("Lanjut?");
    tanya = input3.next().charAt(0);
    if(tanya == 'N'){
        break;
    }
    ++i;
}while(true);
System.out.println("-----------------Hasil DO-WHILE----------------------");
System.out.println("Angka Terkecil : " + terkecil);
System.out.println("Angka Terbesar : " + terbesar);
System.out.println("Total bilangan : " + total);
System.out.println("rata-rata bilangan : " + avg);
```
