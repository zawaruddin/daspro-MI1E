### Percobaan 4: Tebak Angka
1. Pada Percobaan 4 ini, kita akan belajar membuat kode untuk menebak angka menggunakan _nested loop_.
2. Pada percobaan ini kita menggunakan library Scanner untuk menangkap input dari keyboard dan Random untuk meng-generate angka secara acak
3. Ketik dan pahami kode program di bawah ini
![Gambar 6](images/img-06.png)

```java
// Tulis Kode program Percobaan 4 Langkah 3 di atas, disini
import java.util.Scanner; 
import java.util.Random; 

Random random = new Random(); 
Scanner input = new Scanner(System.in); 
char menu = 'y'; 

do{
    int number = random.nextInt(10) + 1; 
    boolean success = false; 
    
    do{
        System.out.print("Tebak angka (1-10) : ");
        int answer = input.nextInt();
        input.nextLine(); 
        
        if (answer == number){
            System.out.println("Yeay.. Tebakan Anda benar... Selamat!!!");
            success = true; 
        }
        
    } while (!success); 
    
        System.out.print("Apakah Anda ingin mengulang permainan (Y/y)?");
        menu = input.nextLine().charAt(0);
    
} while (menu == 'y' || menu == 'Y'); 
```

#### Pertanyaan

2. Apa yang harus dilakukan untuk tidak melanjutkan (tidak mengulangi) permainan tersebut?

    2. hentikan _nested loop_ jika pengguna gagal menebak angka sampai 10x tebakan, dan beri pesan "Maaf Anda gagal menebak angka sebanyak 10x"

#### Jawaban
1. Program diatas merupakan kode untuk menebak angka menggunakan nested loop. Program menggunakan library Random dan bertipe data integer. Sehingga method Random akan menghasilkan nilai integer yaitu dalam rentang -2,147,486,648 sampai +2,147,483,647. Namun parameter dapat ditambahkan dengan cara pergeseran dengan menambahkan nilai pada shifting. Sesuai pada contoh kode program pada percobaan, jika ingin menambahkan parameter 1-10, yaitu caranya int number = random.nextInt(10) + 1; Karena nilai default adalah 0, sehingga parameter tersebut yang awalnya 0-9 akan ditambah dengan 1. Sehingga menghasilkan parameter 1-10. 
2. Untuk tidak melanjutkan (tidak mengulangi) permainan tersebut user cukup mengetikkan huruf selain y pada pertanyaan "Apakah Anda ingin mengulang permainan (Y/y)?" Maka permainan akan berhenti. 


```java
/* Jawaban untuk Percobaan 4 Pertnyaan 3.A disini */
import java.util.Scanner; 
import java.util.Random; 

Random random = new Random(); 
Scanner input = new Scanner(System.in); 
char menu = 'y';

do{
    int number = random.nextInt(10) + 1; 
    boolean success = false; 
    
    do{
        System.out.print("Tebak angka (1-10) : ");
        int answer = input.nextInt();
        input.nextLine(); 
        
        if (answer == number){
            System.out.println("Yeay.. Tebakan Anda benar... Selamat!!!");
            success = true; 
        } else if (answer < number){
            System.out.println("Nilai yang Anda masukkan lebih kecil dari nilai Random!");
            success = false; 
        } else {
            System.out.println("Nilai yang Anda masukkan lebih besar dari nilai Random!");
            success = false; 
        }
        
    } while (!success); 
        
        System.out.print("\nApakah Anda ingin mengulang permainan (Y/y)?");
        menu = input.nextLine().charAt(0);
    
} while (menu == 'y' || menu == 'Y'); 
```

```java
/* Jawaban untuk Percobaan 4 Pertnyaan 3.B disini */
import java.util.Scanner; 
import java.util.Random; 

Random random = new Random(); 
Scanner input = new Scanner(System.in); 
char menu = 'y'; 
int menebak = 0; 

do{
    int number = random.nextInt(10) + 1; 
    boolean success = false; 
    
    do{
        menebak++; 
        System.out.print("Tebak angka (1-10) : ");
        int answer = input.nextInt();
        input.nextLine(); 
        
        if (menebak == 10){
            System.out.print("Maaf Anda gagal menebak angka sebanyak 10x");
            break; 
        } else if (answer == number){
            System.out.println("Yeay.. Tebakan Anda benar... Selamat!!!");
            success = true; 
        } else if (answer < number){
            System.out.println("Nilai yang Anda masukkan lebih kecil dari nilai Random!");
            success = false; 
        } else {
            System.out.println("Nilai yang Anda masukkan lebih besar dari nilai Random!");
            success = false; 
        }
        
    } while (!success); 
    
        System.out.print("\nApakah Anda ingin mengulang permainan (Y/y)?");
        menu = input.nextLine().charAt(0);
    
} while (menu == 'y' || menu == 'Y'); 
```

```java
/* Jawaban Soal 2 disini */

import java.util.Scanner; 

Scanner input = new Scanner(System.in); 

System.out.print("Masukkan nilai N : "); 
int N = input.nextInt();

for (int i = 1; i <= N; i++){
    for (int j = 1; j <= N; j++){
        if (i == 1 || j == 1 || i == N || j == N){
            System.out.print(N);
        } else {
            System.out.print(" ");
        }
    }
    System.out.println(); 
}
```

### Soal 3
Buatlah program untuk mencetak tampilan piramida * seperti gambar di bawah ini, tinggi piramida berdasarkan input dari _keyboard_ N (nilai N minimal 3). Contoh N = 3, dan N = 5
![Soal 3](images/Soal-03.png)

```java
/* Jawaban Soal 3 disini */
import java.util.Scanner;

Scanner input = new Scanner(System.in);
int N; 

System.out.print("Masukkan tinggi segitiga (minimal 3) : ");
N = input.nextInt(); 

for(int i = 1; i <= N; i++){
    for(int j = i; j <= N - 1; j++){
        System.out.print(" ");
    }
    
    for(int k = 0; k <= (i*2)-2; k++){
        System.out.print("*");
    }
    System.out.println(); 
}
```
